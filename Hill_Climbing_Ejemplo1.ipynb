```python
# Importamos las librerías para el algoritmo de Hill Climbing
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
# Definición de la función objetivo
def objetivo(x):
    return -2*x**2 + 4*x + 10
```

```python
# Algoritno de Hill Climbing
def hill_climbing_con_traza(funcion_objetivo, paso=0.1, max_iter=1000):
    x_inicial = random.uniform(-1.5, 3.5)
    x = x_inicial
    traza = [(x, funcion_objetivo(x))]  # Guardar el punto inicial
    print("Valor inicial de X:", x)

    for i in range(max_iter):
        vecino_1 = x + paso
        vecino_2 = x - paso
        mejor_vecino = max(vecino_1, vecino_2, key=funcion_objetivo)

        if funcion_objetivo(mejor_vecino) > funcion_objetivo(x):
            x = mejor_vecino
            traza.append((x, funcion_objetivo(x)))  # Agregar el nuevo punto a la traza
            print("Nuevo valor de x:", x)
        else:
            break

    return x_inicial, x, traza
```

```python
# Ejecutar el algoritmo con traza
x_inicial, resultado, traza = hill_climbing_con_traza(objetivo)

# Crear los datos para la gráfica de la función
x_vals = np.linspace(-1.5, 3.5, 400)
y_vals = objetivo(x_vals)

# Graficar la función y los puntos por los que pasa el algoritmo
plt.figure(figsize=(10, 6))
plt.plot(x_vals, y_vals, label='f(x) = -2x^2 + 4x + 10')

# Graficar cada punto de la traza
for punto in traza:
    plt.scatter(*punto, color='orange')

plt.scatter(x_inicial, objetivo(x_inicial), color='red', label=f'Inicial: x={x_inicial:.2f}')
plt.scatter(resultado, objetivo(resultado), color='green', label=f'Final: x={resultado:.2f}')
plt.title('Optimización con Hill Climbing y traza de la búsqueda')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Definición de la función objetivo
def objetivo2(x):
    return -4*x + 4

# Ejecutar el algoritmo con traza
x_inicial, resultado, traza = hill_climbing_con_traza(objetivo)

# Crear los datos para la gráfica de la función
x_vals = np.linspace(-1.5, 3.5, 400)
y_vals = objetivo(x_vals)
y_vals2 = objetivo2(x_vals)
x_vals0 = x_vals * 0

# Graficar la función y los puntos por los que pasa el algoritmo
plt.figure(figsize=(10, 6))
plt.plot(x_vals, y_vals, label='f(x) = -2x^2 + 4x + 10')
plt.plot(x_vals, y_vals2, label='f(x) = -4x + 4')
plt.plot(x_vals, x_vals0, label='División y = 0', linestyle='dashed', color='black')

# Graficar cada punto de la traza
for punto in traza:
    plt.scatter(*punto, color='orange')

plt.scatter(x_inicial, objetivo(x_inicial), color='red', label=f'Inicial: x={x_inicial:.2f}')
plt.scatter(resultado, objetivo(resultado), color='green', label=f'Final: x={resultado:.2f}')
plt.title('Optimización con Hill Climbing y traza de la búsqueda')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.show()
```
